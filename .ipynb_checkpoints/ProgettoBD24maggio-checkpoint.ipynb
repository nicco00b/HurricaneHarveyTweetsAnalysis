```python
#QUERY 2
#Selezionare i top n hashtag più usati. 
#TODO
#Eventuali miglioramenti: restituire i top n hashtags per giorno o per categoria
def top_n_hashtag(dataset,n):
    hashtag = dataset.select("entities").selectExpr("transform(entities.hashtags,x -> upper(x.text)) as vettore_hashtags")

    hashtag_exploded = hashtag.select(explode(col("vettore_hashtags")))

    result = hashtag_exploded.groupBy("col").agg(count("col")).orderBy(desc(count("col"))).limit(n)
    result = result.withColumnRenamed("col","hashtags").withColumnRenamed("count(col)","hashtag_count")
    return  result

topHashtag = top_n_hashtag(dataset_completo,10).show(truncate = False)
```

```python
#Metodo preliminare che rimuove i tag degli utenti, gli hashtag, i link, caratteri speciali,punteggiatura e stop word.
def pre_processing_text(dataset):
#Parole che iniziano con @ (nomi utente),parole che iniziano con # (hashtag), URL (che iniziano con http:// o https://).
#Simboli, caratteri di controllo, nuove linee e punteggiatura,la stringa RT (indicativo di un retweet),
#caratteri non ASCII (come emoji e caratteri di altre lingue).
    regex = r"\B@\w+\b|\B#\w+\b|https?://\S+|[\p{So}\p{Cntrl}\n\p{Punct}]+|RT|[^\x00-\x7F]+"
    #Sostituiamo tutte le occorrenze che rispettano la regex con uno spazio e rendiamo tutte le parola lower case
    cleaned = dataset.withColumn("clean_text", regexp_replace(col("text"), regex, " "))
    cleaned_text_df = cleaned.withColumn("clean_text_lower", lower(col("clean_text")))

    #crea, per ogni tupla, una lista di parole, il cui separatore è lo spazio
    tokenizer = RegexTokenizer(inputCol="clean_text", outputCol="words", pattern="\\W+")
    dataset_tokenized = tokenizer.transform(cleaned_text_df)

    #rimuove le stop words dalla colonna tokenizzata ( articoli, congiunzioni ecc..)
    remover = StopWordsRemover(inputCol="words", outputCol="words_filtered")
    result = remover.transform(dataset_tokenized)
    return result
```

```python
#QUERY 3
#Le parola più ricorrente per ogni categoria

def max_word_for_category(dataset):
    new_dataset = dataset.withColumn("text",when(col("truncated") == True,"extended_tweet.text").otherwise(col("text")))
    dataset_processato = pre_processing_text(new_dataset)
    
    #dal dataset processato prendiamo la colonna AIDRLabel e la colonna words_filtered esplosa, rinominata come words
    text_with_label = dataset_processato\
                    .select(col("AIDRLabel"), explode(col("words_filtered"))\
                    .alias("words")) #\
    
    #per ogni coppia label-parola, contiamo quante volte occorre la coppia e inseriamo il valore in una nuova colonna words_count                            
    count_words = text_with_label\
                    .groupBy(col("AIDRLabel"),col("words"))\
                    .agg(count("*")\
                    .alias("words_count"))
    
    #per ogni label prendiamo il massimo della colonna words count
    max_word = count_words\
                .groupBy(col("AIDRLabel")\
                .alias("labels"))\
                .agg(F.max("words_count")\
                .alias("max_counted"))
   
    #facciamo la join con la tabella precedente per prendere la parola associata alla coppia label-words_count
    #eliminiamo eventuali parole con conteggio uguale, nella stessa label, e ordiniamo in ordine decrescente
    result = max_word\
            .join(count_words,(max_word.labels == count_words.AIDRLabel) & (max_word.max_counted == count_words.words_count) , "inner")\
            .drop_duplicates(["labels"])\
            .orderBy(desc("max_counted"))
    
    #rimuoviamo le colonne supeflue
    result = result.drop("labels").drop("max_counted")
    
    return result

max_word_for_category(dataset_completo).show(truncate = False)
```

```python

#QUERY 4
#vediamo il trend totale delle parole nei tweet e come una geenrica categoria si adatta a questo trend
def process_text_data(dataset, labels, n=10):
    dataset_processato = pre_processing_text(dataset)

    prova = dataset_processato.select(col("AIDRLabel"), explode(col("words_filtered")).alias("words"))\
            .groupBy("words")\
            .agg(count("words")\
            .alias("words_count"),*[count(when(col("AIDRLabel") == label,1)).alias(label) for label in labels])\
            .orderBy(desc("words_count"))\
            .limit(n)
    
    return prova
   

prova = process_text_data(dataset_completo,[]).show(truncate=False)
```

```python
#QUERY 5 --> giorno, lista utenti, lista location, lista categorie
def top_user_location_categories(dataset, n=5):
    ds_data_convertita = dataset.withColumn("timestamp", from_unixtime(unix_timestamp(col("created_at"), "EEE MMM dd HH:mm:ss ZZZ yyyy")))
    ds_data_AMG = ds_data_convertita.withColumn("data_convertita", date_format(col("timestamp"), "yyyy-MM-dd"))

    #top utenti
    top_utenti = ds_data_AMG.groupBy("data_convertita", "user.name").agg(count("*").alias("num_tweet_utente")).orderBy(desc("num_tweet_utente")).limit(n)
    
    
    return top_utenti 
    
top_user_location_categories(dataset_completo).show(truncate = False)
```

```python
#QUERY 6
def top_mentioned_for_category(dataset, category, n=5):
    #per ogni tupla, prendere il campo entities.user_mention (che restituisce un array di account menzionati) 
    #scandire l'array ed estrarne il nome, infine contare il numero di volte che appare ogni nome
    ds_esploso = dataset.select("AIDRLabel", explode(col("entities.user_mentions")).alias("user_mentions")).filter(col("AIDRLabel") == category) #ogni riga possiede la categoria e un utente mezionato
    ds_users = ds_esploso.select("AIDRLabel", "user_mentions.name")
    risultato = ds_users.groupBy("AIDRLabel", "name").agg(count("*").alias("occorrenze")).orderBy(desc("occorrenze")).limit(n)
    return risultato

ds = top_mentioned_for_category(dataset_completo, "donation_and_volunteering")
ds.show(truncate = False)
```

```python
#QUERY 7  --> categoria, giorno, numero di tweet 
def tweets_distribution_for_category(dataset, category):
    ds_data_convertita = dataset.withColumn("timestamp", from_unixtime(unix_timestamp(col("created_at"), "EEE MMM dd HH:mm:ss ZZZ yyyy")))
    ds_data_AMG = ds_data_convertita.withColumn("data_convertita", date_format(col("timestamp"), "yyyy-MM-dd"))
    risultato = ds_data_AMG.groupBy("AIDRLabel", "data_convertita").agg(count("*").alias("num_tweet_per_giorno"))#.filter(col("AIDRLabel") == category)
    return risultato

ds = tweets_distribution_for_category(dataset_completo, "donation_and_volunteering")
ds.show(truncate = False)
```

```python
#QUERY 8
#tra tutti gli utenti che hanno tweettato almeno una volta nelle date comprese nel nostro dataset
#prendiamo gli utenti più influenti in base al numero di follower, il numero totale di tweet e se il profilo è verificato
 
def most_influents_users(dataset,categories,n=10):
    filtered_df = dataset.filter(col("AIDRLabel").isin(*categories))
    dataset_user = filtered_df.select("user","AIDRLabel")\
    .filter(col("AIDRLabel") != "not_related_or_irrelevant")\
    .filter(col("user.verified") == True)
 
    result = dataset_user.select("AIDRLabel","user.name","user.id","user.followers_count","user.statuses_count")\
                   .dropDuplicates(["id"])\
                   .orderBy(desc("followers_count")).limit(n)
    return result

ds = most_influents_users(dataset_completo, aidr_label_classification)
ds.show(truncate = False)
```

```python
#QUERY 9 
def media_for_category(dataset):
    dataset_tot = dataset.groupBy("AIDRLabel").agg(count("*").alias("occorrenze_totali"))
    dataset_filtrato = dataset.filter(col("entities.media").isNotNull()).groupBy("AIDRLabel").agg(count("*").alias("occorrenze_con_media"))
    dataset_combinato = dataset_tot.join(dataset_filtrato, dataset_filtrato.AIDRLabel == dataset_tot.AIDRLabel, "left")
    risultato = dataset_combinato.withColumn("percentuale %",  round((col("occorrenze_con_media") / col("occorrenze_totali")) * 100, 2)).na.fill(0, "occorrenze_con_media").na.fill(0, "percentuale %")
    return risultato.select(dataset_tot["AIDRLabel"], "occorrenze_con_media", "occorrenze_totali", "percentuale %")

media_for_category(dataset_completo).show(truncate = False)
    
```
